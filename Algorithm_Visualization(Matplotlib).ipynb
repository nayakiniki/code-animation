```python

# Simple Sorting Algorithm Visualization (Bubble Sort)
import matplotlib.pyplot as plt
import numpy as np
import time

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                plt.bar(range(len(arr)), arr)
                plt.pause(0.1)
                plt.clf()
    plt.bar(range(len(arr)), arr)
    plt.show()

arr = np.random.randint(1, 50, 10)
bubble_sort(arr)


```

```python
# Simple Sorting Algorithm Visualization (Quick Sort)
import matplotlib.pyplot as plt
import numpy as np
import time

def quick_sort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quick_sort(arr, low, pi - 1)
        quick_sort(arr, pi + 1, high)

def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] < pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
        visualize(arr)
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    visualize(arr)
    return i + 1

def visualize(arr):
    plt.bar(range(len(arr)), arr)
    plt.pause(0.2)
    plt.clf()

arr = np.random.randint(1, 50, 10)
plt.ion()  # Enable interactive mode for live updates
quick_sort(arr, 0, len(arr) - 1)
plt.bar(range(len(arr)), arr)
plt.show()
```

```python
# Linear Search Animation
import matplotlib.pyplot as plt
import numpy as np
import time

def linear_search(arr, target):
    for i in range(len(arr)):
        plt.bar(range(len(arr)), arr, color=['red' if j == i else 'blue' for j in range(len(arr))])
        plt.pause(0.5)
        plt.clf()
        if arr[i] == target:
            plt.bar(range(len(arr)), arr, color=['green' if j == i else 'blue' for j in range(len(arr))])
            plt.show()
            return i
    plt.bar(range(len(arr)), arr, color='gray')  # If target not found
    plt.show()
    return -1

arr = np.random.randint(1, 50, 10)
target = np.random.choice(arr)  # Pick a random target from the array
plt.ion()  # Enable interactive mode
linear_search(arr, target)
```

```python
# Animating Binary search
import matplotlib.pyplot as plt
import numpy as np
import time

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        plt.bar(range(len(arr)), arr, color=['yellow' if j == mid else 'blue' for j in range(len(arr))])
        plt.pause(0.5)
        plt.clf()
        if arr[mid] == target:
            plt.bar(range(len(arr)), arr, color=['green' if j == mid else 'blue' for j in range(len(arr))])
            plt.show()
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    plt.bar(range(len(arr)), arr, color='gray')  # If target not found
    plt.show()
    return -1

arr = np.sort(np.random.randint(1, 50, 10))  # Generate and sort array
target = np.random.choice(arr)  # Pick a random target
plt.ion()  # Enable interactive mode
binary_search(arr, target)
```

```python
# Animating Depth First Search
import networkx as nx
import matplotlib.pyplot as plt
import time

def visualize(graph, pos, visited, current):
    colors = ['green' if node in visited else 'red' if node == current else 'blue' for node in graph.nodes()]
    plt.clf()
    nx.draw(graph, pos, with_labels=True, node_color=colors, edge_color="black", node_size=800, font_size=16)
    plt.pause(1)

def dfs(graph, node, visited, pos):
    if node not in visited:
        visited.add(node)
        visualize(graph, pos, visited, node)
        for neighbor in graph.neighbors(node):
            dfs(graph, neighbor, visited, pos)

# Define the graph
G = nx.Graph()
edges = [(0, 1), (0, 2), (1, 3), (1, 4), (2, 5), (2, 6)]
G.add_edges_from(edges)
pos = nx.spring_layout(G)  # Position nodes for visualization

# Start DFS from node 0
plt.ion()
dfs(G, 0, set(), pos)
plt.show()
```

```python
# Animating Breadth First Search
import networkx as nx
import matplotlib.pyplot as plt
import time
from collections import deque

def visualize(graph, pos, visited, queue):
    colors = ['green' if node in visited else 'red' if node in queue else 'blue' for node in graph.nodes()]
    plt.clf()
    nx.draw(graph, pos, with_labels=True, node_color=colors, edge_color="black", node_size=800, font_size=16)
    plt.pause(1)

def bfs(graph, start, pos):
    visited = set()
    queue = deque([start])

    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.add(node)
            visualize(graph, pos, visited, queue)
            queue.extend(neighbor for neighbor in graph.neighbors(node) if neighbor not in visited)

# Define the graph
G = nx.Graph()
edges = [(0, 1), (0, 2), (1, 3), (1, 4), (2, 5), (2, 6)]
G.add_edges_from(edges)
pos = nx.spring_layout(G)  # Position nodes for visualization

# Start BFS from node 0
plt.ion()
bfs(G, 0, pos)
plt.show()
```

```python
#Animating dijkstra algorithm
import networkx as nx
import matplotlib.pyplot as plt
import time
import heapq

def visualize(graph, pos, visited, current, distances):
    colors = ['green' if node in visited else 'red' if node == current else 'blue' for node in graph.nodes()]
    plt.clf()
    nx.draw(graph, pos, with_labels=True, node_color=colors, edge_color="black", node_size=800, font_size=16)

    labels = {node: f"{distances[node]}" for node in graph.nodes()}
    nx.draw_networkx_labels(graph, pos, labels=labels, font_color='white', font_size=12)

    plt.pause(1)

def dijkstra(graph, start, pos):
    distances = {node: float('inf') for node in graph.nodes()}
    distances[start] = 0
    visited = set()
    priority_queue = [(0, start)]  # Min-heap (distance, node)

    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)

        if current_node in visited:
            continue

        visited.add(current_node)
        visualize(graph, pos, visited, current_node, distances)

        for neighbor in graph.neighbors(current_node):
            weight = graph[current_node][neighbor]['weight']
            new_distance = current_distance + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                heapq.heappush(priority_queue, (new_distance, neighbor))

# Define a weighted graph
G = nx.Graph()
edges = [(0, 1, 4), (0, 2, 1), (1, 3, 1), (2, 1, 2), (2, 3, 5), (3, 4, 3)]
G.add_weighted_edges_from(edges)
pos = nx.spring_layout(G)  # Position nodes for visualization

# Start Dijkstra from node 0
plt.ion()
dijkstra(G, 0, pos)
plt.show()
```
