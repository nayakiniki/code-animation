```python
%%writefile bubble_sort_animation.py

from manim import *

class BubbleSortAnimation(Scene):
    def bubble_sort(self, numbers, squares, labels):
        n = len(numbers)
        for i in range(n):
            for j in range(0, n - i - 1):
                if numbers[j] > numbers[j + 1]:
                    # Swap values
                    numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]

                    # Swap positions of squares
                    self.play(
                        squares[j].animate.move_to(squares[j + 1].get_center()),
                        squares[j + 1].animate.move_to(squares[j].get_center()),
                        labels[j].animate.move_to(squares[j + 1].get_center()),
                        labels[j + 1].animate.move_to(squares[j].get_center()),
                        run_time=0.5
                    )
                    squares[j], squares[j + 1] = squares[j + 1], squares[j]
                    labels[j], labels[j + 1] = labels[j + 1], labels[j]

    def construct(self):
        numbers = [5, 2, 9, 1, 6]
        squares = []
        labels = []

        # Create number blocks
        for i, num in enumerate(numbers):
            square = Square(color=WHITE).shift(RIGHT * i)
            label = Text(str(num)).move_to(square.get_center())
            squares.append(square)
            labels.append(label)
            self.add(square, label)

        self.wait(1)
        self.bubble_sort(numbers, squares, labels)
        self.wait(2)
```

```python
%%writefile selection_sort_manim.py
from manim import *

class SelectionSortAnimation(Scene):
    def selection_sort(self, numbers, squares, labels):
        n = len(numbers)
        for i in range(n):
            min_idx = i
            for j in range(i + 1, n):
                if numbers[j] < numbers[min_idx]:
                    min_idx = j

            if min_idx != i:
                numbers[i], numbers[min_idx] = numbers[min_idx], numbers[i]

                # Swap animation
                self.play(
                    squares[i].animate.move_to(squares[min_idx].get_center()),
                    squares[min_idx].animate.move_to(squares[i].get_center()),
                    labels[i].animate.move_to(squares[min_idx].get_center()),
                    labels[min_idx].animate.move_to(squares[i].get_center()),
                    run_time=0.8
                )
                squares[i], squares[min_idx] = squares[min_idx], squares[i]
                labels[i], labels[min_idx] = labels[min_idx], labels[i]

    def construct(self):
        numbers = [64, 25, 12, 22, 11]
        squares = [Square(color=WHITE).scale(0.8).shift(RIGHT * i) for i in range(len(numbers))]
        labels = [Text(str(numbers[i]), font_size=36).move_to(squares[i].get_center()) for i in range(len(numbers))]

        for square, label in zip(squares, labels):
            self.add(square, label)

        self.selection_sort(numbers, squares, labels)
        self.wait(2)

```

```python
%%writefile insertion_sort_manim.py
from manim import *

class InsertionSortAnimation(Scene):
    def insertion_sort(self, numbers, squares, labels):
        n = len(numbers)
        for i in range(1, n):
            key = numbers[i]
            j = i - 1

            # Highlight the selected key element
            self.play(labels[i].animate.set_color(YELLOW), run_time=0.5)

            while j >= 0 and numbers[j] > key:
                numbers[j + 1] = numbers[j]
                self.play(
                    squares[j + 1].animate.move_to(squares[j].get_center()),
                    labels[j + 1].animate.move_to(squares[j].get_center()),
                    run_time=0.5
                )
                squares[j + 1], squares[j] = squares[j], squares[j + 1]
                labels[j + 1], labels[j] = labels[j], labels[j + 1]
                j -= 1

            numbers[j + 1] = key
            self.play(labels[i].animate.set_color(WHITE), run_time=0.5)

    def construct(self):
        numbers = [5, 3, 8, 4, 2]
        squares = [Square(color=WHITE).scale(0.8).shift(RIGHT * i) for i in range(len(numbers))]
        labels = [Text(str(numbers[i]), font_size=36).move_to(squares[i].get_center()) for i in range(len(numbers))]

        for square, label in zip(squares, labels):
            self.add(square, label)

        self.insertion_sort(numbers, squares, labels)
        self.wait(2)
```

```python
%%writefile quick_sort_manim.py
from manim import *

class QuickSortAnimation(Scene):
    def partition(self, numbers, squares, labels, low, high):
        pivot = numbers[high]
        i = low - 1

        for j in range(low, high):
            if numbers[j] < pivot:
                i += 1
                numbers[i], numbers[j] = numbers[j], numbers[i]

                # Swap animation
                self.play(
                    labels[i].animate.move_to(squares[j].get_center()),
                    labels[j].animate.move_to(squares[i].get_center()),
                    run_time=0.5
                )
                labels[i], labels[j] = labels[j], labels[i]

        numbers[i + 1], numbers[high] = numbers[high], numbers[i + 1]
        self.play(
            labels[i + 1].animate.move_to(squares[high].get_center()),
            labels[high].animate.move_to(squares[i + 1].get_center()),
            run_time=0.5
        )
        labels[i + 1], labels[high] = labels[high], labels[i + 1]

        return i + 1

    def quick_sort(self, numbers, squares, labels, low, high):
        if low < high:
            pi = self.partition(numbers, squares, labels, low, high)

            self.quick_sort(numbers, squares, labels, low, pi - 1)
            self.quick_sort(numbers, squares, labels, pi + 1, high)

    def construct(self):
        numbers = [10, 7, 8, 9, 1, 5]
        squares = [Square(color=WHITE).scale(0.8).shift(RIGHT * i) for i in range(len(numbers))]
        labels = [Text(str(numbers[i]), font_size=36).move_to(squares[i].get_center()) for i in range(len(numbers))]

        for square, label in zip(squares, labels):
            self.add(square, label)

        self.quick_sort(numbers, squares, labels, 0, len(numbers) - 1)
        self.wait(2)
```

```python
%%writefile linear_search_manim.py
from manim import *

class LinearSearchAnimation(Scene):
    def construct(self):
        numbers = [3, 7, 9, 12, 15, 19, 21]
        target = 15  # Number to search
        squares = [Square(color=WHITE).scale(0.8).shift(RIGHT * i) for i in range(len(numbers))]
        labels = [Text(str(numbers[i]), font_size=36).move_to(squares[i].get_center()) for i in range(len(numbers))]

        for square, label in zip(squares, labels):
            self.add(square, label)

        self.wait(1)

        # Perform Linear Search
        for i in range(len(numbers)):
            self.play(squares[i].animate.set_color(YELLOW), run_time=0.5)

            if numbers[i] == target:
                self.play(squares[i].animate.set_color(GREEN), run_time=0.5)
                break
            else:
                self.play(squares[i].animate.set_color(RED), run_time=0.5)

        self.wait(2)
```

```python
%%writefile binary_search_manim.py
from manim import *

class BinarySearchAnimation(Scene):
    def construct(self):
        numbers = [2, 4, 6, 8, 10, 12, 14, 16]
        target = 10  # Number to search
        squares = [Square(color=WHITE).scale(0.8).shift(RIGHT * i) for i in range(len(numbers))]
        labels = [Text(str(numbers[i]), font_size=36).move_to(squares[i].get_center()) for i in range(len(numbers))]

        for square, label in zip(squares, labels):
            self.add(square, label)

        self.wait(1)

        low, high = 0, len(numbers) - 1

        while low <= high:
            mid = (low + high) // 2
            self.play(squares[mid].animate.set_color(YELLOW), run_time=0.5)

            if numbers[mid] == target:
                self.play(squares[mid].animate.set_color(GREEN), run_time=0.5)
                break
            elif numbers[mid] < target:
                for i in range(low, mid + 1):
                    self.play(squares[i].animate.set_color(RED), run_time=0.5)
                low = mid + 1
            else:
                for i in range(mid, high + 1):
                    self.play(squares[i].animate.set_color(RED), run_time=0.5)
                high = mid - 1

        self.wait(2)
```
